```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import csv
import json
option = webdriver.ChromeOptions()
#option.add_argument("user-data-dir=C:\\Users\\PC2\\AppData\\Local\\Google\\Chrome\\User Data")#to open a window with your default profile (aka your google acc)
driver=webdriver.Chrome(executable_path="C:\\Users\PC2\\Downloads\\chromedriver_win32 (3)\\chromedriver.exe",chrome_options=option)
season_id = [8228,8618,9075]
data = []
for i in season_id:
    data_season = []
    driver.get("https://1xbet.whoscored.com/Regions/252/Tournaments/2/Seasons/{}/England-Premier-League".format(i))
    driver.maximize_window()
    team_stats = driver.find_element(By.XPATH,"//a[contains(text(),'Team Statistics')]")
    team_stats.click()
    for scroll in range(2):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    soup = BeautifulSoup(driver.page_source,"lxml")

    summary_t =soup.find("table",attrs ={"id":"top-team-stats-summary-grid"})
    header_row = summary_t.find('thead')
    columns = [i.text for i in header_row.findAll("th")]
    rows = summary_t.findAll("tr")
    for row in rows[1:]:
        l=[]
        name = row.findAll("td")[0].find("a").text
        l.append(name)
        l.extend([i.text for i in row.findAll("td")[1:3]])
        discipline = [i.text for i in row.findAll("td")[3].findAll("span")]
        l.append(discipline)
        l.extend([i.text for i in row.findAll("td")[4:7]])
        l.append(row.findAll("td")[-1].find("span").text)
        data_season.append(dict(zip(columns,l)))
    data.append(data_season)
final = [{"season":"2020-2021","stats":data[0]},{"season":"2021-2022","stats":data[1]},{"season":"2022-2023","stats":data[2]}]    
with open('whoscored.json', 'w') as outfile: 
        json.dump(final, outfile,indent=0) 





```
