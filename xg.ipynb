```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import csv
import json
option = webdriver.ChromeOptions()
import pandas as pd
```

```python
season_data = {}
for season in [2020,2021,2022]:

  driver=webdriver.Chrome(executable_path="C:\\Users\PC2\\Downloads\\chromedriver_win32 (3)\\chromedriver.exe",chrome_options=option)
  driver.get("https://understat.com/league/EPL/{}".format(season))
  soup = BeautifulSoup(driver.page_source,'lxml')
  scripts = soup.findAll("script")
# Find data for teams
  for el in scripts:
      if 'teamsData' in str(el):
        string_with_json_obj = str(el).strip()
      
# print(string_with_json_obj)

# strip unnecessary symbols and get only JSON data
  ind_start = string_with_json_obj.index("('")+2
  ind_end = string_with_json_obj.index("')")
  json_data = string_with_json_obj[ind_start:ind_end]

  json_data = json_data.encode('utf8').decode('unicode_escape')
  data = json.loads(json_data)
  teams = {i:v for i,v in zip(data.keys(),[data[j]["title"] for j in data.keys()])}
  columns = data[list(data.keys())[0]]["history"][0].keys()
  dataframes = {}
  for id, team in teams.items():
    teams_data = []
    for row in data[id]['history']:
      teams_data.append(list(row.values()))
    
    df = pd.DataFrame(teams_data, columns=columns)
    dataframes[team] = df
    print('Added data for {},season:{}.'.format(team,season))
    season_data[season]=dataframes
  #with open('xg_per_game.json', 'w') as outfile: 
        #json.dump(season_data, outfile,indent=0) 
```
